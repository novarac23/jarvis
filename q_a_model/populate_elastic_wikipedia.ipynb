```python
import os
import json 
from tqdm.notebook import tqdm
from elasticsearch import Elasticsearch
```

```python
config = {'host':'localhost', 'port':9200}
es = Elasticsearch([config])

# test connection
es.ping()
```

```python
index_config = {
    "settings": {
        "analysis": {
            "analyzer": {
                "stop_stem_analyzer": {
                    "type": "custom",
                    "tokenizer": "standard",
                    "filter":[
                        "lowercase",
                        "stop",
                        "snowball"
                    ]
                    
                }
            }
        }
    },
    "mappings": {
        "dynamic": "strict", 
        "properties": {
            "document_title": {"type": "text", "analyzer": "stop_stem_analyzer"},
            "document_text": {"type": "text", "analyzer": "stop_stem_analyzer"}
            }
        }
    }
```

```python
#es.indices.delete(index='wiki-full')
```

```python
es.indices.create(index='wikipedia-full', body=index_config, ignore=400)
```

```python
sub_folders = os.listdir('wiki_dump/text/')
failed_imports = []


for i, folder in enumerate(tqdm(sub_folders)):
    print(f'-----------------folder: {folder}------------------')
    
    for wiki in os.listdir("wiki_dump/text/" + folder):
        wiki_file_path = "wiki_dump/text/" + folder + "/" + wiki
        print(f'we are currently on file {wiki_file_path}')
        
        with open(wiki_file_path, 'rb') as raw_wiki_file:
            tmp = raw_wiki_file.read().decode('utf-8')
            dump = json.dumps(tmp)
            wiki_articles = json.loads(dump)
            wiki_articles = wiki_articles.splitlines(True)
            for i, article in enumerate(wiki_articles):
                try:
                    json_v = json.loads(article)
                    es_json = {}
                    es_json['document_title'] = json_v['title']
                    es_json['document_text'] = json_v['title'] + '' + json_v['text']
                    #es.index(index='wiki-full', id=i, body=es_json)
                    es.index(index='wikipedia-full', body=es_json)
                except Exception as e:
                    failed_imports.append(wiki_file_path)
                    print(f'we could not index article under {wiki_file_path}')
                    
                
        raw_wiki_file.close()
        print(f'so far we have populated {es.count(index="wikipedia-full")["count"]} items in wiki-full.')
```
